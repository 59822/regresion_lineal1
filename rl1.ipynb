## Error del modelo ECM
#### Tenemos que minimizar el error de predicción, por lo cual hacemos una resta de *valor de predicción - valor real*. Podemmos tomar las diferencias del error de predicción para ver cuál modelo es mejor


### Error Cuadrático Medio (ECM)

El **Error Cuadrático Medio (ECM)** se define como:

$ \text{ECM} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 $

Donde:
- $ n $ es el número de observaciones.
- $ y_i $ es el valor verdadero de la $ i $-ésima observación.
- $ \hat{y}_i $ es el valor predicho por el modelo para la $ i $-ésima observación.

El ECM es una medida de la calidad de un modelo de predicción; cuanto menor sea el ECM, mejor será la precisión del modelo.
 el ECM, mejor será la precisión del modelo.

#### Los parametros pára estimar
$ \hat{y}_i = \hat{w}* x_i + \hat{b} $

**Los parametros para estimar**

$\hat{w}$

$\hat{b}$

```python
#Vamos a hacer las predicciones

train['pred'] = train['GrLivArea'] * w + b

#Calculamos el error cuadrático medio
train["diferencia"] = train['SalePrice'] - train['pred']
train["error_cuadratico"] = train["diferencia"]**2
train.head()
```

Cuando buscamos varios **w** vamos a encontrar que se convierte en una función convexa, lo importante de esto es que, entre menor el valor que adopte esa función convexa va a ser el **w** adecuado, entonces tenemos que buscar ese mínimo

```python
# Calcular el error cuadrático medio. Y por tanto
# escoger el w y b que minimicen el error cuadrático medio
w = np.linspace(50, 200, 50)
# Convertimos el array en un dataFrame para poderlo visulizar
grid_error = pd.DataFrame(w, columns = ['w'])
grid_error.head()
```

```python
# le aplicamos la función del minimo cuadrado 

def sum_error(w, train):
    b = 0
    train['pred'] = train['GrLivArea'] * w + b
    train["diferencia"] = train['SalePrice'] - train['pred']
    train["error_cuadratico"] = train["diferencia"]**2
    return(train["error_cuadratico"].mean())
```

```python
grid_error['error'] = grid_error['w'].apply(lambda x: sum_error(x, train = train))
#apply aplica la función a cada elemento de la columna. 
#Lambda x: es una función anónima que recibe un argumento x.

grid_error.head()
```

```python
# Hacemos la grafica de la curva de w
grid_error.plot(x = 'w', y = 'error')
# Toma la columna w y la de error del dataFrame grid_error
plt.show()
```

```python
# Encuentra el índice del valor mínimo en la columna 'error'
min_error_index = grid_error['error'].idxmin()

# Obtén el valor correspondiente de 'w' en ese índice
min_w_value = grid_error.loc[min_error_index, 'w']
#loc es para localizar un valor en un dataFrame
# donde toma el valor del índice min_error_index
# y devuelve el valor de la columna 'w'



print(f"El valor de w que minimiza el error es: {min_w_value}")
```

```python
from sklearn.linear_model import LinearRegression

X_train = np.array(train['GrLivArea']).reshape(-1,1)
y_train = np.array(train['SalePrice']).reshape(-1,1)
# La remoldea a un array de una sola columna

# Crear el modelo
model = LinearRegression(fit_intercept=False)

# Entrenar el modelo
model.fit(X_train, y_train)

# Obtener el valor de w y b
print(f"El valor de la pendiente w es: {model.coef_}")
print(f"El valor del intercepto b es: {model.intercept_}")
```

### 2. Optimización numerica
Donde usamos **el gradiente descendente** donde tomamos un valor arbitrario y calculamos el gradiente, en que dirección que yo mueva los parametros vamos a disminuir los costes. Hasta encontrar el minimo que permite que converg

```python
train.plot.scatter(x='GrLivArea',y='SalePrice')
plt.plot(X_train, model.predict(X_train), color='black')
# plt.plot donde X_train es el valor de x y model.predict(X_train) es el valor de y

plt.show()
```
